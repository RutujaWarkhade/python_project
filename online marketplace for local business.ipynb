```python
import tkinter as tk
from tkinter import Tk, Label, Entry, Button, Text
from tabulate import tabulate  
import pandas as pd
products = {
    "Laptop": {
        "HP": 50000,
        "Lenovo": 60000,
        "DELL": 70000
    },
    "SmartPhone": {
        "Redmi": 10000,
        "Oppo": 15000,
        "Samsung": 12000

    },
    "HeadPhone": {
        "Sony": 150,
        "Bose": 200,
        "Xiomi": 450
    },
    "TV": {
        "Samsung": 26000,
        "SONY": 30000,
        "Xiomi": 25000
    },
}

def view_products():
    def show_products():
        category_choice = category_var.get()
        if category_choice in products:
            product_list.delete('1.0', tk.END)  
            product_list.insert(tk.END, f"Available products in {category_choice}:\n")
            for item, price in products[category_choice].items():
                product_list.insert(tk.END, f"{item} - Rs {price}\n")
        else:
            product_list.delete('1.0', tk.END)  
            product_list.insert(tk.END, "Invalid category!")

    root = tk.Tk()
    root.title("View Products")

    category_label = tk.Label(root, text="Categories:")
    category_label.grid(row=0, column=0, sticky="w")

    category_var = tk.StringVar(root)
    category_var.set(list(products.keys())[0])  

    category_menu = tk.OptionMenu(root, category_var, *products.keys())
    category_menu.grid(row=0, column=1, sticky="w")

    view_button = tk.Button(root, text="View Products", command=show_products)
    view_button.grid(row=1, column=0, columnspan=2)

    product_list = tk.Text(root, height=10, width=40)
    product_list.grid(row=2, column=0, columnspan=2)

    root.mainloop()

def read_excel():
    df = pd.read_excel("C:/Users/OM/Downloads/excel file.xlsx")
    print(df)
    
def make_purchase_gui(shop_name, shop_address, shop_mobile, shop_email):
    cart = {}
    total_price = 0

    def add_to_cart():
        nonlocal total_price
        item = item_entry.get()
        if item.lower() == 'done':
            item_entry.config(state='disabled')
            quantity_entry.config(state='disabled')
            add_to_cart_button.config(state='disabled')
            show_bill_button.config(state='normal')
            name_entry.config(state='normal')
            email_entry.config(state='normal')
            phone_entry.config(state='normal')
            return
        found = False
        for category, items in products.items():
            if item in items:
                found = True
                try:
                    quantity = int(quantity_entry.get())
                    if quantity > 0:
                        cart[item] = cart.get(item, 0) + quantity
                        total_price += products[category][item] * quantity
                    else:
                        feedback_label.config(text="Invalid quantity!")
                except ValueError:
                    feedback_label.config(text="Quantity should be a number!")
                break
        if not found:
            feedback_label.config(text="Item not found!")
        update_cart()

    def update_cart():
        cart_text.delete('1.0', tk.END)
        for item, quantity in cart.items():
            cart_text.insert(tk.END, f"{item} - Quantity: {quantity}\n")

    def show_bill():
        nonlocal total_price
        nonlocal cart
        nonlocal shop_name, shop_address, shop_mobile, shop_email
        total_price_label.config(text=f"Total Price: Rs {total_price}")
        discount = 0
        if total_price > 1000:
            discount = total_price * 0.1
            total_price -= discount
        bill_data = [
            ["Shop Details:", ""],
            ["Name", shop_name],
            ["Address", shop_address],
            ["Mobile", shop_mobile],
            ["Email", shop_email],
            ["", ""], 
            ["Customer Name", name_entry.get()],
            ["Email", email_entry.get()],
            ["Phone Number", phone_entry.get()],
            ["", ""], 
            ["Product", "Quantity", "Price"]
        ]
        for item, quantity in cart.items():
            price = products[next((category for category, items in products.items() if item in items), "")][item] * quantity
            bill_data.append([item, quantity, f"Rs {price}"])

        bill_data.extend([
            ["", ""],  
            ["Total Price without Discount", f"Rs {total_price + discount}"],
            ["Discount (10%)", f"Rs {discount}"],
            ["Total Price with Discount", f"Rs {total_price}"]
        ])
        bill_text.delete('1.0', tk.END)
        bill_text.insert('end', tabulate(bill_data, headers="firstrow", tablefmt="fancy_grid"))

    root = tk.Tk()
    root.title("Make Purchase")

    item_label = tk.Label(root, text="Enter item to purchase (or 'done' to finish):")
    item_label.pack()

    item_entry = tk.Entry(root)
    item_entry.pack()

    quantity_label = tk.Label(root, text="Enter quantity:")
    quantity_label.pack()

    quantity_entry = tk.Entry(root)
    quantity_entry.pack()

    add_to_cart_button = tk.Button(root, text="Add to Cart", command=add_to_cart)
    add_to_cart_button.pack()

    feedback_label = tk.Label(root, text="")
    feedback_label.pack()

    cart_text = tk.Text(root, height=10, width=40)
    cart_text.pack()

    name_label = tk.Label(root, text="Enter your name:")
    name_label.pack()
    name_entry = tk.Entry(root, state='disabled')
    name_entry.pack()

    email_label = tk.Label(root, text="Enter your email:")
    email_label.pack()
    email_entry = tk.Entry(root, state='disabled')
    email_entry.pack()

    phone_label = tk.Label(root, text="Enter your phone number:")
    phone_label.pack()
    phone_entry = tk.Entry(root, state='disabled')
    phone_entry.pack()

    show_bill_button = tk.Button(root, text="Show Bill", command=show_bill, state='disabled')
    show_bill_button.pack()

    total_price_label = tk.Label(root, text="")
    total_price_label.pack()

    bill_text = tk.Text(root, height=20, width=60)
    bill_text.pack()

    root.mainloop()
    
def give_feedback():
    def submit_review():
        review = entry.get()
        if review:
            message.config(text="Thank you for your review!")
        else:
            message.config(text="Please enter a review.")

    root = tk.Tk()
    root.title("Feedback")

    label = tk.Label(root, text="Enter your review:")
    label.pack()

    entry = tk.Entry(root)
    entry.pack()

    submit_button = tk.Button(root, text="Submit", command=submit_review)
    submit_button.pack()

    message = tk.Label(root, text="")
    message.pack()

    root.mainloop()

def main():
    print("Welcome to our online marketplace!")
    while True:
        print("\n1. View Products (press 1):")
        print("2. To see our products details (press 2):")
        print("3. To BUY our Products (press 3):")
        print("4. Give Feedback (press 4):")
        print("5. Exit (press 5)")
        choice = input("Enter your choice: ")

        if choice == '1':
            view_products()
        elif choice=='2':
            read_excel()
        elif choice == '3':
            make_purchase_gui("OM ELECTRONICS", "Rahuri", "9134555590", "omelctronics@mail.com")
        elif choice == '4':
            give_feedback()
        elif choice == '5':
            print("Thank you for visiting our marketplace!")  
            break
        else:
            print("Invalid choice!")

if __name__ == "__main__":
    main()


```
